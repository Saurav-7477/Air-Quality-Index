# Data Collection

```python
import os
import time
import sys
import requests

def data_collection():
    for year in range(2013, 2019):  # for loop for year range
        for month in range(1, 13):  # for loop for month range
            if month < 10:
                # Condition for if month number is below 10
                url = 'https://en.tutiempo.net/climate/0{}-{}/ws-431280.html'.format(month, year)
            else:
                # Condition for 10 to 12 months
                url = 'https://en.tutiempo.net/climate/{}-{}/ws-431280.html'.format(month, year)

            collected_texts = requests.get(url)  # using requests to get the HTML data into collected_texts variable
            collected_text_utf = collected_texts.text.encode('utf-8')  # our HTML contains various data types, so we use utf-8 encoding

            # after getting the data, we need to store it in a directory, so we create the Html_data directory with the year directory
            if not os.path.exists("Data_collection/Html_data/{}".format(year)):
                os.makedirs("Data_collection/Html_data/{}".format(year))

            # To store the data, we need to open the directory
            with open("Data_collection/Html_data/{0}/{1}.html".format(year, month), 'wb') as Result:
                Result.write(collected_text_utf)

            sys.stdout.flush()

if __name__ == "__main__":
    
    data_collection()  # function calling
    
```

# HTML Scapping

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import sys
from bs4 import BeautifulSoup
import os
import csv

def html_scraping(month, year):
    # To scrape the data, we need to give the path of files
    file_path = open('Data_collection/Html_data/{}/{}.html'.format(year, month), 'rb')
    
    # After scraping the data, we need to store it in a variable
    scraped_data = file_path.read()
    
    # Now I create two empty lists for future purpose
    sample_data = []
    final_data = []
    
    # Now I initialize the BeautifulSoup class: BeautifulSoup(scraped_text, file_type)
    soup = BeautifulSoup(scraped_data, 'lxml')
    
    # We need the table data from HTML, so we loop through the table tag and its class from the scraped HTML data
    for table in soup.findAll('table', {'class': 'medias mensuales numspan'}):
        
        # In table data, we need the body of the table to find the features, so we loop through the table body
        for tbody in table:
            
            # In table body, we need the rows to get feature data, so we loop through the table rows
            for tr in tbody:
                
                # Now we extract the row data
                extracted_data = tr.get_text()
                
                # Now we append the row data into the sample_data list
                sample_data.append(extracted_data)
    
    # If we manually check in the HTML, we have 15 features, so to check if we are getting 15 features or not
    # Number of rows
    no_rows = len(sample_data) // 15
    
    # Now to get the feature values, first we need to go through the rows, so we loop through the rows
    for iterate in range(round(no_rows)):
        
        # Creating an empty list to store feature values of each row
        lst = []
        
        # We loop through the features to get each value
        for i in range(15):
            # We add each row data to the empty list
            lst.append(sample_data[0])
        
        # Now we remove data from sample_data
        sample_data.pop(0)
        
        # Now we add the each row values of 15 features to the final_data list
        final_data.append(lst)
    
    length_final_data = len(final_data)
    final_data.pop(length_final_data - 1)
    final_data.pop(0)
    
    # Now we remove the empty features from the table because these features don't contain any value
    for feature in range(len(final_data)):
        final_data[feature].pop(6)
        final_data[feature].pop(13)
        final_data[feature].pop(12)
        final_data[feature].pop(11)
        final_data[feature].pop(10)
        final_data[feature].pop(9)
        final_data[feature].pop(0)
    
    return final_data

# Once scraping the HTML table data,
# HTML table features are independent variables,
# Hours_to_Day PM2.5 feature is a dependent feature,
# we need to combine both independent features and dependent features in a CSV file,
# for that we write a function
def combine_dependent_independent(year, cs):
    for i in pd.read_csv('Data_collection/Html_scraping_data/real_{}.csv'.format(year), chunksize=cs,encoding='latin1'):
        df = pd.DataFrame(data=i)
        mylist = df.values.tolist()
        return mylist

if __name__ == "__main__":
    # We need to create a directory to store the CSV files
    if not os.path.exists("Data_collection/Html_scraping_data"):
        os.makedirs("Data_collection/Html_scraping_data")
    
    # After creating the directory, we need to write the CSV files
    # We need years from 2013 to 2018, so we create a loop for the years
    for year in range(2013, 2019):
        final_data = []
        with open("Data_collection/Html_scraping_data/real_{}.csv".format(year), 'w') as csvfile:
            writing_csv = csv.writer(csvfile, dialect='excel')
            writing_csv.writerow(['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM', 'PM2.5'])
            
            # To add the data to the CSV files, we call the html_scraping function
            for month in range(1, 13):
                temp = html_scraping(month, year)
                final_data += temp
                
            # To get PM2.5 avg values, we need to call the corresponding function
            # For that, we dynamically write it with getattr
            dependent = getattr(sys.modules[__name__], 'Day_values_{}'.format(year))()
            
            # To add the dependent feature PM2.5 to the independent features
            for i in range(len(final_data) - 1):
                final_data[i].insert(8, dependent[i])
            
            with open('Data_collection/Html_scraping_data/real_{}.csv'.format(year), 'a') as csvfile:
                wr = csv.writer(csvfile, dialect='excel')
                for row in final_data:
                    flag = 0
                    for elem in row:
                        if elem == "" or elem == "-":
                            flag = 1
                    if flag != 1:
                        wr.writerow(row)
    
    # We call the combine_dependent_independent function to combine both data
    data_2013 = combine_dependent_independent(2013, 600)
    data_2014 = combine_dependent_independent(2014, 600)
    data_2015 = combine_dependent_independent(2015, 600)
    data_2016 = combine_dependent_independent(2016, 600)
    data_2017 = combine_dependent_independent(2017, 600)
    data_2018 = combine_dependent_independent(2018, 600)
    
    # Combining all years' data into a single CSV
    total = data_2013 + data_2014 + data_2015 + data_2016 + data_2017 + data_2018
    
    with open(r'C:\Users\saura\.jupyter\Real_Combine.csv', 'w') as csvfile:
    # File operations here

        wr = csv.writer(csvfile, dialect='excel')
        wr.writerow(['T', 'TM', 'Tm', 'SLP', 'H', 'VV', 'V', 'VM', 'PM 2.5'])
        wr.writerows(total)
        
df = pd.read_csv(r'C:\Users\saura\.jupyter\Real_Combine.csv', encoding='latin1')

# with open(r'C:\Users\saura\.jupyter\Real_Combine.csv', 'w') as csvfile:
    # File operations here
```

```python
import pickle
import streamlit as st

def load_model():
    pickle_in = open("XGboost.pkl", "rb")
    model = pickle.load(pickle_in)
    return model

def predict_AQI(model, Average_Temperature, Maximum_Temperature, Minimum_Temperature, Atm_pressure_at_sea_level, Average_wind_speed):
    prediction = model.predict([[Average_Temperature, Maximum_Temperature, Minimum_Temperature,
                                 Atm_pressure_at_sea_level, Average_wind_speed]])
    return prediction

def main():
    st.title("Hyderabad AQI Prediction")
    html_temp = """
    <div style="background-color:green;padding:20px">
    <h2 style="color:white;text-align:center;">AQI Prediction ML App</h2>
    </div>
    """
    st.markdown(html_temp, unsafe_allow_html=True)

    model = load_model()

    Average_Temperature = st.text_input("Average Temperature", "Type Here")
    Maximum_Temperature = st.text_input("Maximum Temperature", "Type Here")
    Minimum_Temperature = st.text_input("Minimum Temperature", "Type Here")
    Atm_pressure_at_sea_level = st.text_input("Atmospheric Pressure at Sea Level", "Type Here")
    Average_wind_speed = st.text_input("Average Wind Speed", "Type Here")

    result = ""
    if st.button("Predict"):
        result = predict_AQI(model, Average_Temperature, Maximum_Temperature, Minimum_Temperature,
                             Atm_pressure_at_sea_level, Average_wind_speed)
        st.success('The predicted AQI is: {}'.format(result[0]))

    if st.button("About"):
        st.text("Let's Learn")
        st.text("Built with Streamlit")

if __name__ == '__main__':
    main()
```
